```python
import ollama
import os
import time
import pandas as pd
from google.colab import drive


drive.mount('/content/drive')

questions_file = "/content/drive/MyDrive/klausimai_testavimui.txt"
output_excel = "/content/drive/MyDrive/rezultatai_kursinis.xlsx"

if os.path.exists(questions_file):
    with open(questions_file, "r", encoding="utf-8") as f:
        questions = [line.strip() for line in f.readlines() if line.strip()]
    print(f"Užkrauta {len(questions)} klausimų.")
else:
    raise FileNotFoundError("Nerastas klausimų failas!")


def testuoti_modeli(model_name):
    print(f"\nPradedamas modelio {model_name} testavimas")

    atsakymai = []
    laikai = []

    for i, q in enumerate(questions, 1):
        print(f"[{i}/{len(questions)}] {q[:40]}...")

        start_time = time.time()
        response = ollama.chat(
            model=model_name,
            messages=[
                {'role': 'system', 'content': 'When answering the question be concise. The answer format has to be plain text. No markdown.'},
                {'role': 'user', 'content': q}
                ],
            options = {
                "temperature": 0
            }
        )
        laikai.append(round(time.time() - start_time, 2))
        atsakymas = response['message']['content']
        atsakymai.append(atsakymas)

    df = pd.DataFrame({
        "Klausimas": questions,
        "Atsakymas": atsakymai,
        "Laikas (s)": laikai
    })

    safe_sheet = model_name.replace(":", "_")[:25]

    if os.path.exists(output_excel):
      with pd.ExcelWriter(output_excel, engine = "openpyxl", mode = "a", if_sheet_exists = "replace") as writer:
        df.to_excel(writer, sheet_name = safe_sheet, index = False)
    else:
      with pd.ExcelWriter(output_excel, engine = "openpyxl") as writer:
        df.to_excel(writer, sheet_name = safe_sheet, index = False)

    print(f"Baigtas {model_name} modelio testavimas.\n")
```

```python
models_info = ollama.list()
modeliai = [m.model for m in models_info.models]

for modelis in modeliai:
  testuoti_modeli(modelis)
```

```python
testuoti_modeli("gemma3:4b")
```
